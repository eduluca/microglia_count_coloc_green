# Program to move overlay images into folder

```python
import os
import shutil

# Microglia Images Folder
microglia_folder = r"C:\Users\edward.luca\Desktop\Microglia Counts"
# Overlay Images Folder
overlay_folder = r"C:\Users\edward.luca\Desktop\overlay_images_only_green"

# Iterate through each subfolder inside Microglia Counts
for folder1 in os.listdir(microglia_folder):
    subfolder_path = os.path.join(microglia_folder, folder1)
    
    # Check if the subfolder is a directory
    if os.path.isdir(subfolder_path):
        # Iterate through the files in the subfolder
        for item in os.listdir(subfolder_path):
            item_path = os.path.join(subfolder_path, item)
            
            # Find the Overlay image
            if item.endswith("Overlay.tif"):
                print(item_path)
                # Extract filename from the path
                filename = os.path.basename(item_path)
                # Destination path for the overlay image
                dest_path = os.path.join(overlay_folder, filename)
                # Move the overlay image to the destination folder
                shutil.move(item_path, dest_path)
```
